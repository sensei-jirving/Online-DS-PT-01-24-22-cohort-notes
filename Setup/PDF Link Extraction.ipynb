```python
data = [['Week','Lecture','Title','Filepath','Links']]
for filename in all_pdfs:
	
	folder = os.path.dirname(filename)
	title = os.path.basename(filename)
	week = re.findall(r"Week_\d{2}",folder)[0]
	lect = re.findall(r"Lect\w*_\d{2}",folder)[0]
	
	pdf = pdfx.PDFx(filename)
	links = pdf.get_references_as_dict()
	
	if 'url' in links:
		link_list = links['url']
		link_list = [l for l in link_list if 'df.info' not in l]
	else:
		link_list = ['']
	data.append([week, lect, title, filename,link_list])
	
```

```python
import pandas as pd
raw_df = pd.DataFrame(data[1:], columns=data[0])
raw_df
```

```python
df  =raw_df.explode('Links')
df = df.set_index(['Week','Lecture'])
df
```

```python
# !pip install tabulate
html = df.to_html()
with open('Links from Lecture Slides.html','w') as f:
	f.write(html)
```

```python
df.to_csv('Links from Lecture Slides.csv')
```

```python
txt = df.reset_index().drop(columns=["Filepath"]).to_markdown()
with open('Links from Lecture Slides.md','w') as f:
	f.write(txt)
```
